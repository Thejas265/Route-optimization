```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()

cities = ["A", "B", "C", "D", "E", "F"]
G.add_nodes_from(cities)

edges = [
    ("A", "B", 4), ("A", "C", 2), ("B", "C", 5),
    ("B", "D", 10), ("C", "D", 3), ("C", "E", 8),
    ("D", "E", 2), ("D", "F", 6), ("E", "F", 7)
]
G.add_weighted_edges_from(edges)

def find_shortest_route(graph, start, end):
    path = nx.dijkstra_path(graph, start, end, weight='weight')
    distance = nx.dijkstra_path_length(graph, start, end, weight='weight')
    return path, distance

source, destination = "A", "F"
shortest_path, total_distance = find_shortest_route(G, source, destination)

pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8,6))

nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=2000, font_size=12, font_weight="bold")
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

path_edges = list(zip(shortest_path, shortest_path[1:]))
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color="red", width=2)

plt.title(f"Shortest Path from {source} to {destination}: {shortest_path} (Distance: {total_distance})", fontsize=12)
plt.show()
```
